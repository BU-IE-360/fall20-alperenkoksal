## Introduction
The aim of this work is the forecasting the total number of newly established firms in Turkey. To do this, time series of total number of newly established firms from 2013 to 2020 is selected and analyzed. A model is created by using some independent variables which affect the behavior of the time series data, some seasonality factors which expalins the seasonal behavior of time series data and some special months such as: coup attempt and corona. With this model missing value of year 2020 which corresponds to twelfth month of year 2020 is predicted. This prediction will be checked when new data is added by Electronic Data Delivery System.

Time series data of exchange rate from dollar to turkish liras, time series data of interest rate on commerical loans and time series data of seasonally adjusted confidence index which have some behavior similar to time series data of total number of newly establised firms can explain the behavior of this time series data. Seen from graphs, There are some peaks that affects the performance of the model which is created by using monthly and yearly factors, interest rate on commerical loans, exchange rate from dollar to turkish liras and seasonally adjusted confidence index. These peaks should be analyzed further to improve the model. Also year can be removed from model since it has singularities.

```r
fit=lm(Total_Number~Trend + moy + Year + Interest_rate +  Exchange_rates + Confidence_Index ,newly_established_firms)
newly_established_firms[,fitted:=fitted(fit)]
newly_established_firms[,residual:=residuals(fit)]
summary(fit)

plot(newly_established_firms$Total_Number,type='l',col=2,xlab = "Index",ylab = "Total number of newly established firms",
    main = "Fitted points on The Time Series Data")
points(fit$fitted)
    
plot(fit$residual,xlab = "Index", ylab = "Residuals", main = "Residuals of Model")
```

```r
newly_established_firms[Date=="2016-07-01",coup := 1]
newly_established_firms[is.na(coup) == T,coup := 0]
newly_established_firms[Date=="2020-04-01" | Date =="2020-05-01",corona := 1]
newly_established_firms[is.na(corona) == T,corona := 0]
```

After the addition of the period of coup attempt and corona to model, Model is improved significantly and is good enough to make predictions for next month.

```r
fit=lm(Total_Number~Trend + moy + Interest_rate +  Exchange_rates + Confidence_Index + as.factor(coup) + as.factor(corona) ,newly_established_firms)
newly_established_firms[,fitted:=fitted(fit)]
newly_established_firms[,residual:=residuals(fit)]
summary(fit)

plot(newly_established_firms$Total_Number,type='l',col=2,xlab = "Index",ylab = "Total number of newly established firms",
    main = "Fitted points on The Time Series Data")
points(fit$fitted)
    
plot(fit$residual,xlab = "Index", ylab = "Residuals", main = "Residuals of Model")
```

```r
t12 <- "2020-12-01"
newly_established_firms <- rbind(newly_established_firms,data.table(Date = as.Date(t12)),fill = TRUE)
newly_established_firms[.N,Interest_rate:= newly_established_firms[.N-1,Interest_rate]]
newly_established_firms[.N,Exchange_rates:= newly_established_firms[.N-1,Exchange_rates]]
newly_established_firms[.N,Confidence_Index:= newly_established_firms[.N-1,Confidence_Index]]
newly_established_firms[.N,Trend:= 96]
newly_established_firms[.N,coup:= 0]
newly_established_firms[.N,corona:= 0]
newly_established_firms[.N,Year:= 2021]
newly_established_firms[.N,moy:= "December"]
newly_established_firms[is.na(fitted)==T,fitted:=predict(fit,newly_established_firms[is.na(fitted)==T])]
tail(newly_established_firms)
newly_established_firms[.N,fitted]
```

```r
cols <- c("predicted" = "orange", "actual" = "blue")
ggplot() + 
  geom_line(data = newly_established_firms, aes(x = Date, y = fitted,color = "predicted")) +  
  geom_line(data = newly_established_firms, aes(x = Date, y = Total_Number,color = "actual")) +
  labs(xlab = "Date",ylab = 'Total number of newly established firms', 
       title = 'Actual vs Predicted Total Number of Newly Established Firms' ) +
  scale_color_manual(values = cols)
```
